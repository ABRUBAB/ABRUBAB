```python

import cv2
import matplotlib.pyplot as plt

# Load the uploaded image in grayscale
image_cv2 = cv2.imread("/content/photo_2024-11-29_18-01-45.jpg", cv2.IMREAD_GRAYSCALE)

# Step 1: Horizontal and vertical flips using cv2
flipped_image_h = cv2.flip(image_cv2, 1)  # Horizontal flip
flipped_image_v = cv2.flip(image_cv2, 0)  # Vertical flip

# Step 2: Display input and flipped images side by side
plt.figure(figsize=(15, 10))


# Step 5: Rotate the vertically flipped image by 90 degrees
rotated_image = cv2.rotate(flipped_image_v, cv2.ROTATE_90_CLOCKWISE)

# Display the vertically flipped and rotated image side by side with the original
plt.figure(figsize=(15, 10))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image_cv2, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Vertically Flipped Image')
plt.imshow(flipped_image_v, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Rotated Image (from Vertically Flipped)')
plt.imshow(rotated_image, cmap='gray')

plt.show()



plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image_cv2, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Horizontally Flipped Image')
plt.imshow(flipped_image_h, cmap='gray')


plt.show()

# Step 3: Define a zoom function that focuses on a tumor-like region (bottom-center)
def zoom_image(image, zoom_factor):
    height, width = image.shape[:2]
    new_height, new_width = int(height / zoom_factor), int(width / zoom_factor)
    top = height - new_height  # Focus on bottom region
    left = (width - new_width) // 2  # Center horizontally

    cropped_image = image[top:top+new_height, left:left+new_width]
    zoomed_image = cv2.resize(cropped_image, (width, height), interpolation=cv2.INTER_LINEAR)
    return zoomed_image

# Apply zoom to the tumor-focused region
zoomed_image = zoom_image(image_cv2, zoom_factor=1.5)

# Step 4: Display the zoomed image alongside the original
plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image_cv2, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Zoomed Image (Tumor-Focused)')
plt.imshow(zoomed_image, cmap='gray')

plt.show()
```
