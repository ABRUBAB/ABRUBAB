```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('/content/Te-me_0013.jpg', 0)  # Load as grayscale

# Gamma Correction
gamma = 2.2  # Adjust gamma value as needed
gamma_corrected = np.array(255*(img / 255) ** gamma, dtype = 'uint8')

# CLAHE
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
clahe_img = clahe.apply(img)

# Averaging Filter
kernel_size = 5  # Adjust kernel size as needed
kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)
avg_filtered = cv2.filter2D(img, -1, kernel)

# Fourier Filtering
img_fft = np.fft.fft2(img)
img_shift = np.fft.fftshift(img_fft)

# Create a filter (e.g., low-pass filter)
rows, cols = img.shape
crow, ccol = rows//2, cols//2
low_pass = np.zeros((rows, cols), np.uint8)
low_pass[crow-30:crow+30, ccol-30:ccol+30] = 1

# Apply the filter to the frequency domain
img_shift = img_shift * low_pass

# Inverse Fourier Transform
img_ishift = np.fft.ifftshift(img_shift)
img_back = np.fft.ifft2(img_ishift)
img_back = np.abs(img_back)

# Display the results
plt.figure(figsize=(15, 10))
plt.subplot(231), plt.imshow(img, cmap='gray'), plt.title('Original Image')
plt.subplot(232), plt.imshow(gamma_corrected, cmap='gray'), plt.title('Gamma Corrected')
plt.subplot(233), plt.imshow(clahe_img, cmap='gray'), plt.title('CLAHE')
plt.subplot(234), plt.imshow(avg_filtered, cmap='gray'), plt.title('Averaging Filter')
plt.subplot(235), plt.imshow(img_back, cmap='gray'), plt.title('Fourier Low-Pass Filter')
plt.show()
```
